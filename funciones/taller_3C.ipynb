```python
import numpy as np
import matplotlib.pyplot as plt

def lista():
    numeros=[]
    ent = int(input('Ingrese el numero de datos'))
    for i in range(0,ent):
        b= input("Ingrese los numero de la lista")
        b=float(b)
        numeros.append(b)
        media=np.mean(numeros)
    return media

print('la media de su lista es:', lista())
```

```python
def absoluto(a):
    cero=a[0]
    for i in a:
        if i == 0:
            cero=0
            a=cero
        if i > 0:
            a=np.abs(a)
    return a

absoluto([-2,-1,0,-3,-4])
    
```

```python
def rectangulo(base,altura):
    area=base*altura
    print('La base de rectangulo es:',base, '\n', 'La altura del rectangulo es:', altura, '\n','el area del rectangulo es:',area)
    return area

rectangulo(10,4)
    
```

```python
def triangulo(lado1,lado2,lado3):
    perimetro=lado1+lado2+lado3
    return perimetro

triangulo(1,2,3)
```
